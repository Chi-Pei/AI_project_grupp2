```python
url1 = 'https://maps.googleapis.com/maps/api/geocode/json?address='
url2 = '&key=AIzaSyAzE6YtFO-WRvKEtR2hy8hf2cVqKbIx-Yk'
```

```python
#loop over the df of missing coordinates to get them from the google geocode api
for i in range(0, len(df)):
    addressStr = df.iloc[i,1]
    area = df.iloc[i,0]
    addressArr = addressStr.split(' ')
    address = '+'.join(addressArr)
    url = url1 + address + ',+' + area + url2

    response = requests.get(url)
    resp_json_payload = response.json()
    #print(resp_json_payload['results'])
    if(resp_json_payload['results'] == []):
        df.iloc[i,16] = "NaN"
        df.iloc[i,17] = "NaN"
    else:
        df.iloc[i,16] = resp_json_payload['results'][0]['geometry']['location']['lat']
        df.iloc[i,17] = resp_json_payload['results'][0]['geometry']['location']['lng']
        print(resp_json_payload['results'][0]['geometry']['location'])
```
